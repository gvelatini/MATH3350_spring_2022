# 1970 Vietnam Draft
On December 1, 1969, a lottery was held to determine the order of the call to military service for the year 1970. This lottery applied to all eligible males born between 1/1/1944 and 12/31/1950 (Ages 25 - 31 at the time of the draft).

```r
draft_numbers <- read.csv('DraftData.csv')
draft_numbers <- draft_numbers[order(draft_numbers$draft_number),]
print(head(draft_numbers))

plot(draft_numbers$sequential_date, draft_numbers$draft_number,
    main = '1970 Vietnam Draft Lottery',
    xlab ='day of birth',
    ylab ='lottery number',
)
```

The first draft number drawn was 258, September 12.  All eligible males with this birth day were the first to be sent to Vietnam that year. The first 195 birth dates selected were called to service.  The last of these dates was day 265 (September 24).
